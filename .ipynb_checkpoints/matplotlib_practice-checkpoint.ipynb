```python
import matplotlib.pyplot as plt 
```

```python
saving = [10,20,30,40,50,100]
month = ["jan","feb","mar","april","may","june"] 
```

```python
plt.plot(month, saving)
plt.xlabel("MONTH")
plt.ylabel("saving ($)") 
plt.title("SAVING IN EACH MONTH")
plt.show() 
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09] 
```

```python
plt.plot(x_axis,y_axis ) 
plt.xlabel("Month") 
plt.ylabel("USD") 
plt.show() 
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
#ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
#plt.plot(x_axis,y_axis )
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="NEW york") 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend() 
```

```python
# Create the plot.
#change the figur size 
#plt.figure(figsize=(20, 10)) 
plt.plot(x_axis, y_axis, marker="*"  ,color="blue", linewidth=2, label='NEW YORK') 

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.xlim("Jan", "June")

plt.grid()
# Add the legend.
plt.legend()  
```

```python
plt.plot(x_axis, y_axis , marker="*" , color="green", linewidth=2 , label="boston") 
plt.xlabel("MONTH")
plt.ylabel("FARE")
plt.title("Pyber fare by month")
plt.ylim(0,45)
plt.grid() 

plt.legend() 
```

```python
ax = plt.axes() 
ax.plot(x_axis, y_axis , marker="*" , color="green", linewidth=2 , label="boston") 
ax.set_xlabel("MONTH")
ax.set_ylabel('FARE') 
ax.set_title("PYPER") 
ax.set_ylim(0, 45) 
ax.grid()
ax.legend() 
```
